```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
dataset['title']= le.fit_transform(dataset['title'])
```

```python
import nltk
```

```python
#nltk.download('stopwords')
```

```python
import re
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
c=[]
for i in range(0,67986):
    review=dataset["title"][i]
    review=re.sub('[^a-zA-Z]',' ' , review)
    review=review.lower()
    review=review.split()
    review=[word for word in review if not word in set(stopwords.words('english'))]
    ps=PorterStemmer()
    review=[ps.stem(word) for word in review if not word in set(stopwords.words('english'))]
    review=' '.join(review)
    c.append(review)
```

```python
c
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=30000)
x=cv.fit_transform(c).toarray()
x=dataset.iloc[:,0:7]
x=dataset.iloc[:,0:7].values
y=dataset.iloc[:,-1].values
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
sc= StandardScaler()
x_train=sc.fit_transform(x_train)
x_test= sc.transform(x_test)
```

```python
import keras
from keras.models import Sequential
from keras.layers import Dense
```

```python
model = Sequential()
model.add(Dense(output_dim= 12, init='uniform', activation= 'relu',input_dim=7))
```

```python
model.add(Dense(output_dim=100,init='uniform',activation='relu'))
```

```python
model.add(Dense(output_dim=3,init='uniform', activation='softmax'))
```

```python
model.compile(optimizer='adam',loss='binary_crossentropy',metrics= ['accuracy'])
```

```python
model.fit(x_train,y_train,epochs=40,batch_size=32)
```

```python
model.save('1stmodel.h5')
```
